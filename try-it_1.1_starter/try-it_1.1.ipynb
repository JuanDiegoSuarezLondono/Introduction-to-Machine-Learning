### In this exercise, generate some basic charts using Pandas. 


```python
# import data from pandas
import pandas as pd
import os

#Make directory called plots
os.makedirs("plots", exist_ok=True)

#create the DataFrame utilizing Pandas 
data = pd.read_csv("https://media.geeksforgeeks.org/wp-content/uploads/nba.csv") 

#generate all info for this dataset
data.info()

#Remove the last line of NAN
data = data.iloc[:-1 , :]

#Ignore any warnings
import warnings
warnings.filterwarnings('ignore')
```

###  Part One: In this exercise, generate a basic histogram.


```python
#Please generate a basic histogram and save the histogram as a png file
# Hint: use pandas histogram function.

ax = data.plot.hist(title="Histogram", grid=True)
ax.figure.savefig('demo-file.pdf')
```

###  Part Two: In this exercise, generate a basic scatter plot. 


```python
#Please generate a basic scatter plot utilizing Age as the x axis and Salary as the y axis
# Hint: use pandas scatter function to generate the chart.

scat = data.plot.scatter(x='Age',y='Salary', title="Age vs Salary", grid=True)
scat = print(scat)


```

###  Part Three: In this exercise, generate a basic lineplot. 

```python
# Please generate a basic line plot by name as the x axis and salary as the y axis 
#for the "Golden State Warriors" using the nba data set. 
#Note a new dataframe called teamdata has been created for you that will be used during the rest of this 



teamdata = data.filter(items=['Team', 'Name', 'Salary'])
teamdata = teamdata.loc[data.Team == 'Golden State Warriors', 'Name':'Salary']
teamdata.set_index('Name')


lines = teamdata.plot.line(x='Name', y='Salary',title="Player vs Salaey",rot=45, grid=True, figsize=(10,10))


# ANSWER CHECK
print(lines)
```

###  Part Four: In this exercise, generate a basic bar chart. 

```python
# Please generate a basic bar plot by name as the x axis and salary as the y axis of the teamdata dataframe.

bar = teamdata.plot.bar(x='Name', y='Salary', xlabel='Name', ylabel='Salary', title="Player vs Salary", rot = 55, grid = True)


print(bar)

```
