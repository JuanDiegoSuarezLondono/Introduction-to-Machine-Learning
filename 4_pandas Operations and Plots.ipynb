<h1 style="color:green"><center>Column operations<center></h1> 

```python
table = pd.DataFrame({'A':[0.1,0.2,0.3],'B':[10,20,30]})
table
```

```python
type(table['A'])
```

A series is, apart from the dataframe, the other user-facing object class in pandas 

# [Series methods](https://pandas.pydata.org/docs/reference/frame.html): sum, multiply, add, division, summation, percentage
Every column of a DataFrame is a pandas Series object, and is equipped with many useful methods that can be used to do computations.
### Example: Sum of a column

```python
table['B'].sum()
```

### Example: Add two columns 

```python
table['C'] = table['A'].add(table['B'])
table
```

### Example: Mulltiply two columns 

```python
table['C'] = table['A'].multiply(table['B'])
table
```

### Division
To divide two numbers in Python you can separate the two values using the / sign.

```python
table['C'] = table['A']/table['B']
table
```

### Summation
To sum two numbers in python you can add the two values together using the plus sign.

```python
table['C'] = table['A']+table['B']
table
```

### Percentage
While pandas does not natively provide a function to calculate the percentage, this can be done by dividing the column by its sum and then multiplying by 100 as shown below:

```python
table['C'] = (table['B']/table['B'].sum())*100
table
```

<h1 style="color:green"><center>Plotting with pandas<center></h1> 

- DataFrame.plot() is an all-purpose tool for plotting data in DataFrames.
- See the documentation for important parameters: x, y, kind, etc.

```python
#Scikit-learn offers several test datasets.
#The iris dataset is a classical dataset for regression
from sklearn import datasets
X = datasets.load_iris(as_frame=True)
df = X.frame
df
```

### kind = 'hist'

```python
df.plot(kind='hist', y=['sepal length (cm)','sepal width (cm)'])
```

### Kind = 'scatter' 

```python
df.plot(kind='scatter', x='sepal length (cm)',y='sepal width (cm)', grid=True)
```

### kind='line'
These are more common for time series data

```python
df = pd.read_csv('DataSets_starter/timeseries.csv')
df
```

```python
df.plot(kind='line',x='Date', y=['Wind','Solar'], grid=True, figsize=(10,10))
```

### kind = 'bar'
