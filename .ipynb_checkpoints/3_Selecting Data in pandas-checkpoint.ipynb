<h1 style="color:green"><center>Selecting Data</center></h1>

```python
import pandas as pd
```

There are three basic ways to select data from a DataFrame
- **table[]:** Bracket selectrion
- **table.loc[]:** Label-based selection
- **table.iloc[]:** Integer-based selection

# [] Bracket selectrion
Syntax: X[selector]
- To select columns the selector can be
    - A list of column label (string)
    - A list of column labels (list of strings)
- To select rows the selector con be
    - A slice
    - A boolean mask
    
### Example DataSet


```python
filename= 'DataSets_starter/celebrity-heights.csv'
CH = pd.read_csv(filename)
CH
```

```python
CH2 = CH.set_index('fullname')
CH2
```

### Select a single column(string)

```python
CH['fullname']
```

### Select multiple columns (list of string) 

```python
CH[['fullname','lastname']]
```

### Select rows with slice indexing
Slice indexing, in which you specify a start index, a stop index, and a step, works in the same way in Panadas as it does for regular Python lists.

```python
#Select the first 4 rows
CH[:4]
```

### Select rows with a boolean mask
You can use a boolean list or a boolean Series as a mask for filtering rows.

```python
"""This returns a list of booleans false for those that 
are shorter than 2 meters"""
CH['meters']>2
```

```python
"""This returns the full data set for those that 
are shorter than 2 meters"""
CH[CH['meters']>2]
```

# loc[] : Label-based selection
Syntax: X.loc[rowselector,columnselector] (columnselector is optional)
rowselector can be
- An index (integer, string, date, etc.)
- A list of indexes.
- A slice (only if the index is integer-based)
- A boolean mask

columnselector can be
- A string label
- A list of strings labels

### .loc can do (almost) anything [] can do
- Except: .loc[slice] when the index is not integer-based

### Example: Select rows with index list 

```python
CH.loc[0:3]
```

```python
CH.loc[3]
```

```python
CH.loc[[0,5,7]]
```

```python
#This data set have a non-numeric index
CH2.loc[['David Rappaport','Warwick Davis']]
```

### Example: Select multiple rows and columns 

```python
CH2.loc[['David Rappaport','Warwick Davis'],['midname','lastname']]
```

### Example: Boolean masking 

```python
CH.loc[CH['meters']>2,['midname','lastname']]
```

# iloc[]: Integer-based selection
Syntax: iloc[rowselector,columnselector] (columnselector is optional)
rowselector and columnselector can be
- An integer or list of integers
- A slice

```python
CH2.iloc[:3,:3]
```
